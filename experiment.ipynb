```python
import pandas as pd

stocks_data = pd.read_csv('data/stocks.csv')
indexes = pd.read_csv('data/indexes.csv')

stocks_data['Date'] = pd.to_datetime(stocks_data['Date'], utc=True)
indexes['Date'] = pd.to_datetime(indexes['Date'], utc=True)

stocks_data.set_index('Date', inplace=True)
indexes.set_index('Date', inplace=True)

stocks_names = [name for name in stocks_data.columns if "_dividends" not in name]
dividends_names = [name for name in stocks_data.columns if "_dividends" in name]

stocks = stocks_data[stocks_names]
dividends = stocks_data[dividends_names]

```

```python
n_stocks = 3

some_stocks = stocks.iloc[:,:n_stocks+1]

print(some_stocks.head())
some_stocks.plot()

some_dividends = dividends.iloc[:, :n_stocks+1]
some_dividends.plot()
```
